### Exploration File

This notebook aims to find out the most optimal duration of a clap sound.

### Import Libraries

```python
%matplotlib inline
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
from pydub import AudioSegment
```

### Load Audio

```python
audio_path = "../audio/mp3/audio2.mp3"

# slice the performance
audio = AudioSegment.from_file(audio_path, format="mp3")
cut_audio = audio[:32000]
cut_audio.export("test.wav", format="wav")

ipd.display(cut_audio)
```

### Visual the Time Domain

```python
claps, sr = librosa.load("test.wav")

librosa.display.waveshow(claps)
plt.show()
```

### Trying different duration

The idea is to find the duration where the audio sounds clear that it is humans clapping before and after a performance

```python
# 1 second
cut_audio = audio[:1000]
ipd.display(cut_audio)
```

```python
# 1.5 seconds
cut_audio = audio[:1500]
ipd.display(cut_audio)
```

```python
# 1.8 seconds
cut_audio = audio[:1800]
ipd.display(cut_audio)
```

```python
# 2 seconds
cut_audio = audio[:2000]
ipd.display(cut_audio)
```

1.8 seconds seem to be the sweet spot. 1 second is too short that it is difficult to recognise that it is the sound of humans clapping hands.
