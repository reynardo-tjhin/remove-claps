### Import Libraries

```python
%matplotlib inline
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
from pydub import AudioSegment
```

### Load Audios

```python
audio_no = 17
audio_path = f"../audio/wav/audio{audio_no}.wav"

start = 2761.08
end = 2795.1

# slice the performance
audio = AudioSegment.from_file(audio_path, format="wav")
audio = audio[start*1000:end*1000]
audio.export("test.wav", format="wav")

ipd.display(audio)
```

### Visualise the Time Domain

```python
claps, sr =librosa.load("test.wav")

# plot the figure
plt.figure(figsize=(50,8))

librosa.display.waveshow(claps)
plt.xticks(np.arange(0, len(claps) / sr, 0.2))
plt.show()
```
