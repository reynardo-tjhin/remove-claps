### Import Libraries

```python
%matplotlib inline
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
from pydub import AudioSegment
```

### Load Audios

```python
audio_path = "../audio/mp3/audio19.mp3"

start = 26
end = 44

# slice the performance
audio = AudioSegment.from_file(audio_path, format="mp3")
cut_audio = audio[start*1000:end*1000]
cut_audio.export("test.wav", format="wav")

ipd.display(cut_audio)
```

### Visualise the Time Domain

```python
claps, sr = librosa.load("test.wav")

librosa.display.waveshow(claps)
plt.show()
```
