### Import Libraries

```python
%matplotlib inline
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

### Load Audio

```python
claps_file = "./audio/test.wav"
claps, sr = librosa.load(claps_file)
```

### Spectral Centroid and Visualisation

```python
FRAME_SIZE = 1024
HOP_LENGTH = 512
claps_sc = librosa.feature.spectral_centroid(y=claps, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]

frames = range(len(claps_sc))
t = librosa.frames_to_time(frames)

plt.figure(figsize=(25, 10))
plt.plot(t, claps_sc, color='b')

plt.show()
```

### Spectral Bandwidth and Visualisation

```python
claps_bw = librosa.feature.spectral_bandwidth(y=claps, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]

plt.figure(figsize=(25, 10))
plt.plot(t, claps_bw, color='b')

plt.show()
```
