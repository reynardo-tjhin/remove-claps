### Sound Power

- Rate at which energy is transferred
- Energy per unit of time emitted by a sound source in all directions
- Measured in watt (W)

### Sound intensity

- Sound power per unit area
- Measured in W/m^2

### Threshold of hearing

- Human can perceive sounds with very small intensities
- Threshold of Hearing (TOH) = 10^-12 W/m^2
- Threshold of Pain (TOP) = 10 W/m^2

### Intensity level

- Logarithmic scale
- Measured in decibels (dB)
- Ration between 2 intensity values
- Use an intensity of reference (TOH)

Equation:
dB(I) = 10 * log (I / I of TOH)

0 dB = Intensity of Hearing

- Every increase of 3 dBs, intensity doubles

### Loudness

- (Sort of like: used to imagine amplitude)
- Subjective perception of sound intensity
- Depends on duration / frequency of a sound
- Depends on age
- Measured in _phons_

### Timbre

- Colour of sound
- Diff between two sounds with same intensity, frequency, duration
- Described with words like: bright, dark, dull, harsh, warm

#### Features of timbre

- Timbre is multidimensional

1. Sound envelope
2. Harmonic content
3. Amplitude / Frequency modulation

##### Features of timbre: Sound envelope

- Attack-Decay-Sustain-Release Model
- For example: a hit on a piano key, the sound goes from a very high amplitude (in the Attack section), it decays then sustains and then is released.

##### Features of timbre: Harmonic content

Before going into harmonic content,
Complex Sound:
- Superposition of sinusoids
- A _partial_ is a sinusoid used to describe a sound
- The lowest partial is called _fundamental frequency_
- A harmonic partial is a frequency that is a multiple of the fundamental frequency
<br> For example: f1 = 440, f2 = 2 * 440, f3 = 3 * 440
- _Inharmonicity_ indicates a deviation from a harmonic partial

##### Features of timbre: Frequency / Amplitude Modulation

Frequency Modulation

- AKA vibrato
- Periodic variation in frequency
- In music, used for expressive purposes

Amplitude Modulation

- AKA tremolo
- Periodic variation in amplitude
- In music, used for expressive purposes
