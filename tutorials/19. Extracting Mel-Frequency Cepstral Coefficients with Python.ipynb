### Import Libraries

```python
%matplotlib inline
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

### Load Audio File

```python
claps_file = "./audio/test.wav"
claps, sr = librosa.load(claps_file)
```

### Extract MFCCs

```python
mfccs = librosa.feature.mfcc(y=claps, n_mfcc=13, sr=sr)

print(f"Shape of MFCCs: {mfccs.shape}")
```

### Visualise MFCCs

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(data=mfccs,
                         x_axis="time",
                         sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

### Calculate delta and delta2 MFCCs

```python
delta_mfccs = librosa.feature.delta(mfccs)
delta2_mfccs = librosa.feature.delta(mfccs, order=2)

print(f"Shape of Delta MFCCs:   {delta_mfccs.shape}")
print(f"Shape of Delta2 MFCCs:  {delta2_mfccs.shape}")
```

### Visualise Delta MFCCs

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(data=delta_mfccs,
                         x_axis="time",
                         sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

### Visualise Delta2 MFCCs

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(data=delta2_mfccs,
                         x_axis="time",
                         sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

### Comprehensive MFCCs

```python
comprehensive_mfccs = np.concatenate((mfccs, delta_mfccs, delta2_mfccs))

print(f"Shape of Comprehensive MFCCs:   {comprehensive_mfccs.shape}")
```

### Visualising Comprehensive MFCCs

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(data=comprehensive_mfccs,
                         x_axis="time",
                         sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```
