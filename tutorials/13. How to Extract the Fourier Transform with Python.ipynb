### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import IPython.display as ipd
```

### Load Audio Files

```python
clapping_file = "./audio/test.wav"

# displaying the audio
ipd.Audio(clapping_file)

# load the file using librosa
claps, sr = librosa.load(clapping_file)
```

### Get the Fourier Transform

```python
claps_ft = np.fft.fft(claps)

print(f"Number of samples:              {len(claps)}")
print(f"Length of Fourier Transformed:  {len(claps_ft)}")

print()

print(f"Example of first item of Fourier Transform: {claps_ft[0]}")
print(f"Magnitude of the first itme:                {abs(claps_ft[0])}")
```

### Visualising the Magnitude Spectrum

```python
magnitude_spectrum = np.abs(claps_ft)

frequency_ratio = 0.1

plt.figure(figsize=(18,5))

frequency = np.linspace(0, sr, len(magnitude_spectrum))
num_frequency_bins = int(len(frequency) * frequency_ratio) # 1 is the frequency ratio

plt.plot(frequency[:num_frequency_bins], magnitude_spectrum[:num_frequency_bins])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")

plt.show()
```
