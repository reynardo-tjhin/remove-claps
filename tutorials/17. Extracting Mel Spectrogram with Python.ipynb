### Import Libraries

```python
%matplotlib inline
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

### Load Audio file

```python
claps_file = "./audio/test.wav"
claps, sr = librosa.load(claps_file)
```

### Mel Filter Banks

```python
filter_banks = librosa.filters.mel(n_fft=2048, sr=sr, n_mels=10)

print(f"Size of filter banks: {filter_banks.shape}")
```

### Visualise Mel Filter Banks

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(data=filter_banks,
                         sr=sr,
                         x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()

# There are 10 mel bands (10 colored blocks below)
```

### Extracting Mel-Spectrogram

```python
mel_spectrogram = librosa.feature.melspectrogram(y=claps, sr=sr, n_fft=2048, hop_length=512, n_mels=90)

print(f"Shape of mel spectrogram: {mel_spectrogram.shape}")
# 90 mel bands

# Visualise
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
plt.figure(figsize=(25, 10))
librosa.display.specshow(data=log_mel_spectrogram,
                         x_axis="time",
                         y_axis="mel",
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
